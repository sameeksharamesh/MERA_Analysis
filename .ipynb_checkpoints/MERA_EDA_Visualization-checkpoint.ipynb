```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/lakshaykapur/Developer/ADS/MERA_Analysis/DMAT/Mera_CleanData.csv")
print("Dataset Description")
print(df.info())
```

```python
df.head()
```

```python
print("Summary Statistics")
df.describe()
```

```python
df['data'] = pd.to_datetime(df['date'])
```

```python
# EDA: Visualizations
eda_figures = {}

# Distribution of target variable: extreme_rain
eda_figures["extreme_rain_dist"] = plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="extreme_rain", palette="viridis")
plt.title("Distribution of Extreme Rain Events")
plt.xlabel("Extreme Rain (0: No, 1: Yes)")
plt.ylabel("Count")

# Correlation heatmap for numerical features
eda_figures["correlation_heatmap"] = plt.figure(figsize=(12, 10))
numerical_features = df.select_dtypes(include=["float64", "int64"])
corr_matrix = numerical_features.corr()
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")

# Distribution of a numerical feature: precipitation
eda_figures["precipitation_dist"] = plt.figure(figsize=(6, 4))
sns.histplot(df["rain"], kde=True, bins=30, color="blue")
plt.title("Distribution of Precipitation")
plt.xlabel("Precipitation")
plt.ylabel("Density")

# Boxplot of precipitation by extreme_rain
eda_figures["precipitation_boxplot"] = plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="extreme_rain", y="rain", palette="Set2")
plt.title("Precipitation vs Extreme Rain")
plt.xlabel("Extreme Rain (0: No, 1: Yes)")
plt.ylabel("Precipitation")

# Pairplot of selected features
selected_features = ["rain", "pressure_cbl", "wind_speed", "temp_range", "extreme_rain"]
eda_figures["pairplot"] = sns.pairplot(df[selected_features], hue="extreme_rain", palette="husl")

eda_figures
```
